```python
import pandas as pd #importing pandas to Python
import plotly.express as px #importing Ploty Express (for the graphs)
```

```python
df_principal = pd.read_excel("/content/Imersão Python - Aula 1 + 2 [Cópia].xlsx", sheet_name="Principal") #getting the path for the file (dataframe) to be read
df_principal.head(10) #getting the first 10 lines of a dataframe | 'df_principal' prints the entire dataframe
```

```python
df_total_acoes = pd.read_excel("/content/Imersão Python - Aula 1 + 2 [Cópia].xlsx", sheet_name="Total_de_acoes") #getting another tab from the dataframe
df_total_acoes #printing the entire dataframe
```

```python
df_ticker = pd.read_excel("/content/Imersão Python - Aula 1 + 2 [Cópia].xlsx", sheet_name="Ticker") #getting another tab from the dataframe
df_ticker #printing the entire dataframe
```

```python
df_chat_gpt = pd.read_excel("/content/Imersão Python - Aula 1 + 2 [Cópia].xlsx", sheet_name="GPT") #getting another tab from the dataframe
df_chat_gpt #printing the entire dataframe
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy() #getting specific columns from dataframe | 'copy()' destroys original dataframe and create a new one
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)': 'valor_final', 'Var. Dia (%)': 'var_dia_pct'}).copy() #changing columns names
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct']/100 #creating new column
df_principal['valor_inicial'] = df_principal['valor_final']/ (df_principal['var_pct'] + 1) ##creating columns and using basic math
df_principal
```

```python
#dataFrame.merge(dataFrameToBeMerged, left_on="column from 'dataFrame'", right_on="column from 'dataFrameToBeMerged'", how="type of join")
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left') #going through dataframes' array [aka VLOOKUP] | Left Outer Join
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código']) #removing column from dataframe
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica'] #creating another column
df_principal
```

```python
pd.options.display.float_format = "{:.2f}".format #formating dataframe to show all float values up until 2 digits
```

```python
df_principal
```

```python
df_principal["Qtde. Teórica"] = df_principal["Qtde. Teórica"].astype(int) #formating column to 'int' type
```

```python
df_principal
```

```python
df_principal = df_principal.rename(columns={"Qtde. Teórica": "qtd_teorica"}).copy() #renaming column to good practices
df_principal
```

```python
#creating new column & applying data based on values | got from ChatGPT
# 'lambda' == foreach()
df_principal['Resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Neutro' if x == 0 else 'Desceu')) # apply() is used to trigger arrow/lambda functions
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on="Ativo", right_on="Ticker", how="left") #merging "Ticker" table into "Principal"
df_principal = df_principal.drop(columns=["Ticker"]) #dropping "Ticker" column, so only "Nome" stays
df_principal
```

```python
df_principal = df_principal.merge(df_chat_gpt, left_on="Nome", right_on="Nome da Empresa", how="left")
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={"Idade (anos)": "idade"}).copy() #changing column name
df_principal
```

```python
df_principal = df_principal.rename(columns={'idade': "Idade"}).copy()
df_principal
```

```python
#made with GPT
df_principal['Status'] = df_principal['Idade'].apply(lambda x: 'Centenária' if x > 100 else ('Mais de meio século' if x > 50 else 'Menos de 50')) #creating new column based on existing one
df_principal
```

```python
#made through GPT
# Encontrando o maior valor na coluna "variacao_rs"
maior = df_principal['variacao_rs'].max()

# Encontrando o menor valor na coluna "variacao_rs"
menor = df_principal['variacao_rs'].min()

# Calculando a média dos valores na coluna "variacao_rs"
media = df_principal['variacao_rs'].mean()

# Calculando a média dos valores positivos na coluna "variacao_rs"
media_positiva = df_principal[df_principal['variacao_rs'] > 0]['variacao_rs'].mean()

# Calculando a média dos valores negativos na coluna "variacao_rs"
media_negativa = df_principal[df_principal['variacao_rs'] < 0]['variacao_rs'].mean()

# Imprimindo os resultados
print("Maior\t", "R$ {:,.2f}".format(maior))  # Imprime o maior valor formatado como moeda
print("Menor\t", "R$ {:,.2f}".format(menor))  # Imprime o menor valor formatado como moeda
print("Média\t", "R$ {:,.2f}".format(media))  # Imprime a média formatada como moeda
print("Média Total Positiva\t", "R$ {:,.2f}".format(media_positiva))  # Imprime a média dos valores positivos formatada como moeda
print("Média Total Negativa\t", "R$ {:,.2f}".format(media_negativa))  # Imprime a média dos valores negativos formatada como moeda
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == "Subiu"] #creating new dataframe based on only "Subiu" values
df_principal_subiu
```

```python
#creating new dataframe | sum() would create a vector, but reset_index() forces it to stay as a dataframe
#new_dataframe = dataFrame.groupby("first column")["column to be attached"].sum() |-> sum() joins every instance into a single summed-up row
df_analise_segmento = df_principal_subiu.groupby("Segmento")["variacao_rs"].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby("Resultado")["variacao_rs"].sum().reset_index()
df_analise_saldo
```

```python
#generating bar graphs through Plotly
# graph = px.bar(dataFrame, x="column to be horizontal", y="column to be vertical", text="text input on 'y'", title="graph title")
fig = px.bar(df_analise_saldo, x="Resultado", y="variacao_rs", text="variacao_rs", title="Variacao Reais por Resultado")
fig.show()
```
